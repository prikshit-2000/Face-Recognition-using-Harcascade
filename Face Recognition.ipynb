```python
img = plt.imread('group.jpg')
plt.imshow(img)
```

```python
img.shape
```

```python
model=cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
faces=model.detectMultiScale(img)
```

```python
faces
```

```python
faces.shape
```

```python
for face in faces:
    x,y,w,h = face
    img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
plt.imshow(img)
```
